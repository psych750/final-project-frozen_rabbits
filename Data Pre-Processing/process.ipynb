```python
import pandas as pd

# Load the datasets
workplace_closures_df = pd.read_csv('/Users/Jun/Desktop/750 project 3/workplace-closures-covid.csv')
population_df = pd.read_csv('/Users/Jun/Desktop/750 project 3/population.csv')
covid_df = pd.read_csv('/Users/Jun/Desktop/750 project 3/WHO-COVID-19-global-data.csv')
gdp_composition_df = pd.read_csv('/Users/Jun/Desktop/750 project 3/gdp_composition_2020.csv')
imports_and_exports_df = pd.read_csv('/Users/Jun/Desktop/750 project 3/trade.csv')
```

Rename the columns so the datasets can be joined by the 'country' column.

```python
# Rename columns so that the datasets can be joined by "country"
workplace_closures_df = workplace_closures_df.rename(columns={'Entity': 'country'})
population_df = population_df.rename(columns={'Country (or dependency)': 'country'})
covid_df = covid_df.rename(columns={'Country': 'country'})
imports_and_exports_df = imports_and_exports_df.rename(columns={'Country': 'country'})
```

The covid_df demonstrates the cumulative cases in each country in each day from 2020-01-03 to 2023-11-30. We can use the data for 2020-12-31 for each country to represent the cumulative cases in 2020.

```python
covid_2020_df = covid_df[covid_df['Date_reported'] == '2020-12-31'][['country', 'Cumulative_cases', 'Cumulative_deaths']]
covid_2020_df.to_csv('/Users/Jun/Desktop/750 project 3/covid_2020_df.csv', index = False)
```

Unify the country names according to the GDP composition dataset. Check if any different names for one country in the five datasets. All the country names in the other 4 datasets are compared with be the same with the GDP composition dataset. Then we corrected them to be the same with the GDP composition dataset in Excel.

```python
# Extracting country names from both datasets
gdp_countries = set(gdp_composition_df['country'])
imports_exports_countries = set(imports_and_exports_df['country'])

# Finding countries that are named differently in the two datasets
# We check if each country in one dataset is not present in the other dataset
different_countries_gdp = gdp_countries - imports_exports_countries
different_countries_imports_exports = imports_exports_countries - gdp_countries

different_countries_gdp, different_countries_imports_exports
```

```python
# Do it again for other 3 datasets
population_countries = set(population_df['country'])

different_countries_gdp = gdp_countries - population_countries
different_countries_population = population_countries - gdp_countries

different_countries_gdp, different_countries_imports_exports
```

```python
covid_countries = set(covid_2020_df['country'])

different_countries_gdp = gdp_countries - covid_countries
different_countries_covid = covid_countries - gdp_countries

different_countries_gdp, different_countries_covid
```

```python
workplace_closures_countries = set(workplace_closures_df['country'])

different_countries_gdp = gdp_countries - workplace_closures_countries
different_countries_workplace_closures = workplace_closures_countries - gdp_countries

different_countries_gdp, different_countries_workplace_closures
```

Merge the data.

```python
merged_df_1 = pd.merge(workplace_closures_df, population_df, on='country')
merged_df_2 = pd.merge(merged_df_1, covid_2020_df, on='country')
merged_df_3 = pd.merge(merged_df_2 , gdp_composition_df, on='country')
merged_df_4 = pd.merge(merged_df_3 , imports_and_exports_df, on='country')
merged_df_4.to_csv('/Users/Jun/Desktop/750 project 3/merged_df.csv', index = False)
```
